```python
top_companies = df.groupby("Company")["Laid_Off"].sum().sort_values(ascending=False).head(10)

top_companies
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
top_companies.plot(kind="bar")

plt.title("Top 10 Companies by Layoffs")
plt.xlabel("Company")
plt.ylabel("Employees Laid Off")
plt.xticks(rotation=45)

plt.show()
```

```python
import plotly.express as px
```

```python
top_companies = df.groupby("Company")["Laid_Off"].sum().sort_values(ascending=False).head(10)

fig = px.bar(
    x=top_companies.index,
    y=top_companies.values,
    labels={"x": "Company", "y": "Employees Laid Off"},
    title="Top 10 Companies by Layoffs"
)

fig.show()
```

```python
continent = df.groupby("Continent")["Laid_Off"].sum().reset_index()

fig = px.pie(
    continent,
    names="Continent",
    values="Laid_Off",
    title="Layoffs by Continent"
)

fig.show()
```

```python
yearly = df.groupby("Year")["Laid_Off"].sum().reset_index()

fig = px.line(
    yearly,
    x="Year",
    y="Laid_Off",
    markers=True,
    title="Year-wise Layoffs"
)

fig.show()
```

```python
map_df = df.dropna(subset=["latitude", "longitude", "Laid_Off"])

fig = px.scatter_geo(
    map_df,
    lat="latitude",
    lon="longitude",
    size="Laid_Off",
    hover_name="Company",
    color="Industry",
    title="Global Tech Layoffs"
)

fig.show()
```

```python
top_funding = df.groupby("Company")["Money_Raised_in__mil"].max().sort_values(ascending=False).head(10)

top_funding
```

```python
plt.figure(figsize=(10,5))

top_funding.plot(kind="bar", color="purple")

plt.title("Top 10 Companies by Funding")
plt.xlabel("Company")
plt.ylabel("Money Raised (Million USD)")
plt.xticks(rotation=45)

plt.show()
```

```python
avg_layoffs = df.groupby("Industry")["Laid_Off"].mean().sort_values(ascending=False).head(10)

avg_layoffs
```

```python
plt.figure(figsize=(10,5))

avg_layoffs.plot(kind="bar", color="teal")

plt.title("Average Layoffs by Industry")
plt.xlabel("Industry")
plt.ylabel("Average Employees Laid Off")
plt.xticks(rotation=45)

plt.show()
```

# Conclusion

This project analyzed global technology layoffs from 2020 to 2025 using Python.

### Key Findings
- The highest layoffs occurred during 2022 and 2023.
- The technology industry experienced the largest number of layoffs.
- North America was the most affected region.
- Several companies laid off 100% of their workforce.
- Higher funding did not always prevent companies from reducing employees.
- Layoff trends varied across industries, countries, and company stages.

This analysis helps understand global workforce trends and the impact of economic conditions on the technology sector.
